# Imports

All necessary imports to label the images for regression

```python
import numpy as np

from torchvision import transforms
from torch.utils.data import Dataset, ConcatDataset

import sharpnesses as sh
import helper
```

# Transforms 

The following transforms are applied to images

```python
TF = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((300, 300)),
    transforms.Normalize((0.5,), (0.5,))
])
```

# Dataset Class

A class with the implemented logic to assign every image in the stack the value of sharpness based on the given function

```python
class RegressionDataset(Dataset):
    def __init__(self, stack, sharpness_func=None, transform=None):
        self.stack = stack
        self.sharpness_func = sharpness_func
        self.transform = transform

        self.data = self.assign_data()
        self.labels = self.assign_labels()

    def assign_data(self):
        return self.stack

    def assign_labels(self):
        sharpness = []
        for i in range(len(self.stack)):
            sharpness.append(self.sharpness_func(self.stack[i]))
        return np.array(sharpness)
                
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        image, label  = self.data[idx], self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
```
