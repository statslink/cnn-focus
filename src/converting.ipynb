# Imports

```python
import scyjava
import imagej

import numpy as np
```

# ImageJ initialization

Along with the additional memory allocation

```python
scyjava.config.add_options('-Xmx8g')

ij = imagej.init()
print(f"ImageJ version: {ij.getVersion()}")
```

# A path for loading

```python
PATH_IN = 'data/'
```

# A path fot saving

```python
PATH_OUT = 'data/'
```

# Loading

## 10x magnification 

```python
dataset_10x_0 = ij.io().open(PATH_IN + '10x_air_gir_05um.tif')
dataset_10x_1 = ij.io().open(PATH_IN + '10x_diff3_05um.tif')
```

## 20x magnification 

```python
dataset_20x_0 = ij.io().open(PATH_IN + '20x_diff3_04um.tif')
```

## 40x magnification 

```python
dataset_40x_0 = ij.io().open(PATH_IN + '40x_diff0_04um.tif')
dataset_40x_1 = ij.io().open(PATH_IN + '40x_diff1_04um.tif')
dataset_40x_2 = ij.io().open(PATH_IN + '40x_diff2_04um.tif')
dataset_40x_3 = ij.io().open(PATH_IN + '40x_diff3_04um.tif')
dataset_40x_4 = ij.io().open(PATH_IN + '40x_diff4_04um.tif')
dataset_40x_5 = ij.io().open(PATH_IN + '40x_diff5_04um.tif')
```

## 60x magnification

```python
dataset_60x_0 = ij.io().open(PATH_IN + '60x_diff3_02um.tif')
```

## 100x magnification 

```python
dataset_100x_0 = ij.io().open(PATH_IN + '100x_IR_02um.tif')
dataset_100x_1 = ij.io().open(PATH_IN + '100x_IRO_02um.tif')
dataset_100x_2 = ij.io().open(PATH_IN + '100x_MWIRO_02um.tif')
```

## memory usage

```python
ij.getApp().getInfo(True)
```

# Converting

## 10x magnification

```python
dataset_10x_0_np = np.array(ij.py.from_java(dataset_10x_0))
dataset_10x_1_np = np.array(ij.py.from_java(dataset_10x_1))
```

```python
dataset_10x_0_np = (dataset_10x_0_np / 256).astype(np.uint8)
dataset_10x_1_np = (dataset_10x_1_np / 256).astype(np.uint8)
```

## 20x magnification

```python
dataset_20x_0_np = np.array(ij.py.from_java(dataset_20x_0))
```

```python
dataset_20x_0_np = (dataset_20x_0_np / 256).astype(np.uint8)
```

## 40x magnification

```python
dataset_40x_0_np = np.array(ij.py.from_java(dataset_40x_0))
dataset_40x_1_np = np.array(ij.py.from_java(dataset_40x_1))
dataset_40x_2_np = np.array(ij.py.from_java(dataset_40x_2))
dataset_40x_3_np = np.array(ij.py.from_java(dataset_40x_3))
dataset_40x_4_np = np.array(ij.py.from_java(dataset_40x_4))
dataset_40x_5_np = np.array(ij.py.from_java(dataset_40x_5))
```

```python
dataset_40x_0_np = (dataset_40x_0_np / 256).astype(np.uint8)
dataset_40x_1_np = (dataset_40x_1_np / 256).astype(np.uint8)
dataset_40x_2_np = (dataset_40x_2_np / 256).astype(np.uint8)
dataset_40x_3_np = (dataset_40x_3_np / 256).astype(np.uint8)
dataset_40x_4_np = (dataset_40x_4_np / 256).astype(np.uint8)
dataset_40x_5_np = (dataset_40x_5_np / 256).astype(np.uint8)
```

## 60x magnification

```python
dataset_60x_0_np = np.array(ij.py.from_java(dataset_60x_0))
```

```python
dataset_60x_0_np = (dataset_60x_0_np / 256).astype(np.uint8)
```

## 100x magnification

```python
dataset_100x_0_np = np.array(ij.py.from_java(dataset_100x_0))
dataset_100x_1_np = np.array(ij.py.from_java(dataset_100x_1))
dataset_100x_2_np = np.array(ij.py.from_java(dataset_100x_2))
```

```python
dataset_100x_0_np = (dataset_100x_0_np / 256).astype(np.uint8)
dataset_100x_1_np = (dataset_100x_1_np / 256).astype(np.uint8)
dataset_100x_2_np = (dataset_100x_2_np / 256).astype(np.uint8)
```
